# Packages

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd

%matplotlib inline
plt.rcParams['font.sans-serif'] = 'Arial' # Choose a nice font here
plt.rcParams['font.size'] = 10 # standard font size
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Arial'
plt.rcParams['mathtext.it'] = 'Arial'
plt.rcParams['figure.facecolor'] = 'white' # make sure background isn't transparent
plt.rcParams['figure.dpi'] = 150
```

# Functions and constants

# Read data

```python
TEmin = pd.read_csv('temin.csv', sep=';')
```

```python

TE   = TEmin['TE']
cpx01  = TEmin['01 cpx']
opx01 = TEmin['01 opx']
cpx03 = TEmin['03 cpx']
cpx04  = TEmin['04 cpx']
opx04 = TEmin['04 opx']
cpx05  = TEmin['05 cpx']
opx05 = TEmin['05 opx']
DM  = TEmin['DM']
```

# Analysis

# Plots

```python
#opx
plt.plot(TE,opx01,c='red', label='TVG 01 opx')
plt.plot(TE,opx04,c='green', label='TVG 04 opx')
plt.plot(TE,opx05,c='yellow', label='TVG 05 opx')
plt.plot(TE,cpx01,c='red', label='TVG 01 cpx')
plt.plot(TE,cpx03,c='blue', label='TVG 03 cpx')
plt.plot(TE,cpx04,c='green', label='TVG 04 cpx')
plt.plot(TE,cpx05,c='yellow', label='TVG 05 cpx')
plt.plot(TE,DM,c='k', label='DM', linewidth=3, linestyle=':')
plt.plot(TE,DM,c='k', label='DM', linewidth=2)
plt.yscale('log')
plt.ylabel('PUM normalised')
plt.legend(loc ='lower right', frameon=False)
```
